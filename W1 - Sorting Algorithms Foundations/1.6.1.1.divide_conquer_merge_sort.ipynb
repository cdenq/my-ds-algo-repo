```python
def helper(arr, start, end):
    #DIVIDE STEP
    #leaf node code
    if start >= end: #if len of subarray is 1
        return
    else: #arbitrary internal node
        mid = (start + end) / 2
        helper(arr, start, mid)
        helper(arr, mid + 1, end)

    #SOLVE STEP
    #we assume they solve

    #COMBINE STEP
    i, j = start, mid + 1
    aux = [] #aux will have final len of (end - start + 1)
    while (i <= mid) & (j <= end):
        if arr[i] <= arr[j]:
            aux.append(arr[i])
            i += 1
        else:
            aux.append(arr[j])
            j += 1

    #when pointers reach end of the subarrays...
    while i <= mid:
        aux.append(arr[i])
        i += 1
    while j <= end:
        aux.append(arr[j])
        j += 1
    
    #cpoy over aux values into arr and return
    arr[start:end+1] = aux
    return arr

def merge_sort(arr):
    helper(arr, 0, len(arr) - 1)
    return arr
```

```python
arr = [5, 6, 1, 3, 5, 1, 2, 6, 9, 3]
print(merge_sort(arr))
```
